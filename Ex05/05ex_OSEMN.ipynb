4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
file_name = "./credit_card.dat"
f = open(file_name)

lines = f.read().split("\n")
ln = len(lines[0]) - 4
size = 6
formatted_lines = []
for line in lines[0:10]:
    temp_line = ""
    for i in range(int(ln/size)):
        esabit=line[i*6:(i+1)*6]
        temp_line += chr(int(esabit, 2))
    formatted_lines.append(temp_line)

formatted_lines
```

5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.

```python
import json

with open("user_data.json", mode='r') as user_data_file:
    user_data = json.load(user_data_file)
    filtered_user_data = [data for data in user_data if data["CreditCardType"] == "American Express"]

with open("filtered_user_data.csv", mode='w') as filtered_user:
    json.dump(filtered_user_data, filtered_user, ensure_ascii='False')
```
